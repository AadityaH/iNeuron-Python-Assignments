```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic=pd.read_csv(url)
titanic=pd.DataFrame(titanic)
titanic.head()
titanic.fillna(0)
titanic.drop(index=[len(titanic)-1],inplace=True)
```

```python
# Plotting Gender Ration in Pie Chart

gender=titanic['sex'].value_counts()
gender.plot.pie(autopct = '%.2f')
```

```python
## Scatterplot with Fare Paid and age with different colour for gender
c=pd.factorize(titanic['sex'])
titanic.plot(kind='scatter',x='age',y='fare',c=c[0],alpha=0.5)
```
